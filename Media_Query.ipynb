```python
""" 1.  Create a simple webpage with a header, a main content section, and a footer. Apply a media query that
changes the background color of the header to blue when the screen width is less than 600px. """

        <!DOCTYPE html>
        <html lang="en">

        <head>
            <meta charset="UTF-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Responsive Header</title>
            <style>
                * {
                    margin: 0;
                    padding: 0;
                }

                header {
                    height: 100px;
                    background-color: whitesmoke;
                    display: flex;
                    justify-content: center;
                    align-items: center;
                }

                ul {
                    display: flex;
                    list-style-type: none;
                }

                a {
                    text-decoration: none;
                    font-size: 25px;
                    padding: 20px;
                    color: black;
                    font-weight: bold;
                }

                @media screen and (max-width:588px) {
                    header {
                        background-color: blue;
                        color: white;
                    }
                }
            </style>
        </head>

        <body>
            <header>
                <ul>
                    <li><a href="">Responsive Header</a></li>
                </ul>
            </header>
            <br>
            <main>
                Main content
            </main>
            <br>
            <footer>
                Footer
            </footer>
        </body>

        </html>
```

```python
""" 3.  Create a navigation bar with five links. Apply a media query that converts the navigation into a vertical list for
screens smaller than 768px and also changes the background color to gray. """

        <!DOCTYPE html>
        <html lang="en">

        <head>
            <meta charset="UTF-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Document</title>
            <style>
                header{
                    height: auto;
                    background-color: black;
                    padding: 10px;
                }

                ul {
                    list-style-type: none;
                    display: flex;
                    justify-content: space-around;
                    align-items: center;
                }
                a {
                    text-decoration: none;
                    font-size: 20px;
                    color: white;
                }

                @media screen and (max-width:648px) {
                    header {
                        background-color: gray;
                        padding: 0;
                    }

                    ul {
                        display: flex;
                        flex-direction: column;
                    }

                    li {
                        padding: 10px;
                    }
                }
            </style>
        </head>

        <body>
            <header>
                <ul>
                    <li><a href="">Home</a></li>
                    <li><a href="">About</a></li>
                    <li><a href="">Services</a></li>
                    <li><a href="">Portfolio</a></li>
                    <li><a href="">Contact</a></li>
                </ul>
            </header>
        </body>

        </html>
```

```python
""" 4.  Create a grid based layout containing four grid items which arranges them as the screen size changes. """

        <!DOCTYPE html>
        <html lang="en">
        <head>
            <meta charset="UTF-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Document</title>
            <style>
                .container{
                    display: grid;
                    grid-template-columns: 1fr 1fr 1fr;
                }
                .box {
                    width: 590px;
                    height: 300px;
                    margin: 20px;
                    color: white;
                    display: flex;
                    justify-content: center;
                }

                .box1 {
                    background-color: red;
                }

                .box2 {
                    background-color: green;
                }

                .box3 {
                    background-color: blue;
                }

                @media screen and (max-width:490px) {
                    .container {
                        display: grid;
                        grid-template-columns: 1fr;
                    }

                    .box{
                        width: 400px;
                    }
                }
            </style>
        </head>
        <body>
            <div class="container">
                <div class="box1 box">Box 1</div>
                <div class="box2 box">Box 2</div>
                <div class="box3 box">Box 3</div>
                <div class="box1 box">Box 4</div>
            </div>
        </body>
        </html>
```

```python
""" 5.  Create the below given layout using the flexbox in css, which should adapt itself on mobile screen as per given
below output. """

        <!DOCTYPE html>
        <html lang="en">
        <head>
          <meta charset="UTF-8" />
          <meta name="viewport" content="width=device-width, initial-scale=1.0" />
          <title>Flex Responsive Layout</title>
          <style>
            * {
          margin: 0;
          padding: 0;
          box-sizing: border-box;
        }

        body {
          font-family: Arial, sans-serif;
        }

        .container {
          height: 100vh;
          display: flex;
          align-items: center;
          justify-content: center;
        }

        .row {
          display: flex;
          flex-direction: row;
          width: 100%;
          height: 100%;
        }

        .column {
          display: flex;
          flex-direction: column;
          flex: 1;
        }

        .box {
          display: flex;
          justify-content: center;
          align-items: center;
          color: white;
          font-size: 20px;
        }

        /* Specific boxes */
        .box1, .box4 {
          width: 80px;
          background-color: #333;
        }

        .box2 {
          background-color: mediumseagreen;
          flex: 1;
        }

        .box3 {
          background-color: darkorange;
          flex: 1;
        }

        .box4 {
          background-color: #333;
        }

        /* Responsive */
        @media (max-width: 768px) {
          .row {
            flex-direction: column;
            align-items: stretch;
          }

          .box1, .box4 {
            width: 100%;
            height: 60px;
          }

          .column {
            flex-direction: column;
          }

          .box2, .box3 {
            height: 100px;
          }
        }

          </style>
        </head>
        <body>
          <div class="container">
            <div class="row">
              <div class="box box1">Box1</div>
              <div class="column">
                <div class="box box2">Box2</div>
                <div class="box box3">Box3</div>
              </div>
              <div class="box box4">Box4</div>
            </div>
          </div>
        </body>
        </html>
```
